```python
data = db.question.find({},{"qtext":1, "qcomment":1, 'is_cook':1})
```

```python
X = []
Y = []
for item in data:
    text = item.get('qtext','')+' '+item.get('qcomment','')
    if text:
        X.append(text)
        Y.append(item.get('is_cook', False))
```

```python
len(X), len(Y)
```

```python
import numpy as np
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
count_vectorizer = CountVectorizer(min_df=3, max_df=0.6)
```

```python
X_counted = count_vectorizer.fit_transform(X)
```

```python
tfidf_vectorizer = TfidfVectorizer(min_df=3, max_df=0.6)
```

```python
X_tfidf = tfidf_vectorizer.fit_transform(X)
```

```python
X_tfidf.shape
```

```python
from sklearn.cross_validation import train_test_split
from sklearn.metrics import classification_report, roc_auc_score
from sklearn.naive_bayes import BernoulliNB, MultinomialNB
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X_counted, Y)
```

```python
clf_bnb = BernoulliNB()
clf_bnb.fit(X_train, Y_train)
Y_pred = clf_bnb.predict(X_test)
Y_pred_proba = clf_bnb.predict_proba(X_test)
print('roc_auc', roc_auc_score(y_true=Y_test, y_score=Y_pred_proba[:,1]))
print(classification_report(y_true=Y_test, y_pred=Y_pred))
```

```python
clf_mnb = MultinomialNB()
clf_mnb.fit(X_train, Y_train)
Y_pred = clf_mnb.predict(X_test)
Y_pred_proba = clf_mnb.predict_proba(X_test)
print('roc_auc', roc_auc_score(y_true=Y_test, y_score=Y_pred_proba[:,1]))
print(classification_report(y_true=Y_test, y_pred=Y_pred))
```

```python
iverted_vocab = {_id:w for (w,_id) in count_vectorizer.vocabulary_.items()}
for _id in clf_bnb.coef_.argsort()[0][-150:]:
    print(iverted_vocab[_id], clf_bnb.coef_[0][_id])
```

```python
iverted_vocab = {_id:w for (w,_id) in count_vectorizer.vocabulary_.items()}
for _id in clf_mnb.coef_.argsort()[0][-150:]:
    print(iverted_vocab[_id], clf_mnb.coef_[0][_id])
```
